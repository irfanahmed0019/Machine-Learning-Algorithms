```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("irfanasrullah/groceries")

print("Path to dataset files:", path)
```

## 2. Load Dataset

```python
import os
print(os.listdir(path))
```

```python
df=pd.read_csv(path+'/groceries - groceries.csv')
```

## 3. Remove Item Count Column

```python
df = df.drop(columns=['Item(s)'])
```

## 4. Convert Rows into Transactions

```python
transactions = []

for i in range(len(df)):

    row = []

    for item in df.iloc[i]:

        if pd.notna(item):
            row.append(str(item))

    transactions.append(row)

print(transactions[:5])
```

## 5. Transaction Encoding

```python
te = TransactionEncoder()

te_array = te.fit(transactions).transform(transactions)

basket = pd.DataFrame(
    te_array,
    columns=te.columns_
)

print(basket.head())
```

##6. FP-Growth

```python

frequent_items = fpgrowth(
    basket,
    min_support=0.02,
    use_colnames=True
)

print(frequent_items.head())
```

##7. Generate Association Rules

```python
rules = association_rules(
    frequent_items,
    metric="confidence",
    min_threshold=0.3
)

print(rules.head())
```
